```python

class MinQueue(object):
    def __init__(self):
        from collections import deque
        self.q = deque()

    def push(self, x):
        count = 1
        while self.q and self.q[-1][0] > x:
            count += self.q.pop()[1]
        self.q.append([x, count])

    def pop(self):
        self.q[0][1] -= 1
        if not self.q[0][1]:
            self.q.popleft()

    def min(self):
        return self.q[0][0]


def solve(n, m, costs):
    q = MinQueue()
    for i in range(n):
        if i > m:
            q.pop()

        c = costs[i] if costs[i] > 0 else float('inf')
        if i == 0:
            q.push(0)
        elif q.q:
            q.push(q.min() + c)

    return q.min() if q.min() < float('inf') else -1

f=open('C:\\Users\\My_NoteBook\\Desktop\\Fb Hacker Cup\\Question 4\\running_on_fumes_chapter_1_sample_input.txt','r')
TC = int(f.readline())
for tc in range(1, TC+1):
    n, m = map(int, f.readline().split())
    costs = [int(f.readline()) for _ in range(n)]

    ans = solve(n, m, costs)
    print(f"Case #{tc}: {ans}")
```
