```python
import pandas as pa
import numpy as np
import seaborn as sn
import matplotlib.pyplot as pt
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
sd=pa.read_csv("E:\ebooks CSE\datasets\BalanceScale.csv",names=['class','rb','lb','rw','lw'])
```

```python
#sd
```

```python
x=sd.values[:,1:]
y=sd.values[:,0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=753)
```

```python
# Decision Tree
clf_entropy= DecisionTreeClassifier(min_samples_split=0.05)
clf_entropy.fit(x_train,y_train)
```

```python
y_pred_en=clf_entropy.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_en)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion="entropy",n_estimators=11,random_state=992)
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_rf)
```

```python
from sklearn.tree import DecisionTreeClassifier
ts_score=[]
for j in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=j)
    lr=DecisionTreeClassifier(min_samples_split=0.05).fit(x_train,y_train)
    ts_score.append(lr.score(x_test,y_test))
print(ts_score.index(np.max(ts_score)))
```

```python
from sklearn.ensemble import RandomForestClassifier
ts_score=[]
for j in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=j)
    lr=RandomForestClassifier(min_samples_split=0.05).fit(x_train,y_train)
    ts_score.append(lr.score(x_test,y_test))
print(ts_score.index(np.max(ts_score)))
```
