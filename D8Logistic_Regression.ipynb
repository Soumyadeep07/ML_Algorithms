```python
td=sd.drop(['Name','Ticket','Cabin','Embarked','Fare','SibSp','Parch'],axis=1)
td.head()
```

```python
sn.boxplot(x='Pclass',y='Age',data=td)
```

```python
def age_approx(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pa.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
td['Age']=td[['Age','Pclass']].apply(age_approx,axis=1)
```

```python
gender=pa.get_dummies(td['Sex'])
```

```python
td.drop(['Sex'],axis=1,inplace=True)
```

```python
tdmy=pa.concat([td,gender],axis=1)
tdmy.head()
```

```python
x=tdmy.iloc[:,[0,2,3,4,5]].values
y=tdmy.iloc[:,1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=417)
```

```python
from sklearn.linear_model import LogisticRegression
ts_score=[]
for j in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=j)
    lr=LogisticRegression().fit(x_train,y_train)
    ts_score.append(lr.score(x_test,y_test))
print(ts_score.index(np.max(ts_score)))
```

```python
logReg=LogisticRegression(solver='lbfgs')
logReg.fit(x_train,y_train)
y_pred=logReg.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
